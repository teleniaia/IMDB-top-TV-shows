```python
import requests
import csv
from bs4 import BeautifulSoup
import pandas as pd
import plotly.express as px

page = requests.get('https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250') # Getting page HTML through request
soup = BeautifulSoup(page.text, 'html.parser') # Parsing content using beautifulsoup
#Pulls specific data from IMDB 
links1 = soup.select("td.titleColumn a")
links2 = soup.select("td.titleColumn span")
links3 = soup.select("td.ratingColumn.imdbRating strong")

mytitles = " "
myyear = " "
myrating = " "

for anchor in links1:
  mytitles += anchor.text + " \n "
for anchor in links2:
  myyear += anchor.text + " \n "
for anchor in links3:
  myrating += anchor.text + " \n "

mytitles_i = mytitles.split("\n")
myyear_i = myyear.split("\n")
myrating_i = myrating.split("\n")

df = pd.DataFrame({'Title' : (mytitles_i), 'Year' : (myyear_i), 'Rating' :(myrating_i)})


data_df = df

fig = px.scatter(data_df, x = "Year" , y = "Rating" , color= "Title" )
fig.update_xaxes(categoryorder='category ascending')


fig.show()

#with open('df.csv', 'w') as f:
   # writer = csv.writer(f)
   # writer.writerows(df)
   # pathout = '/home/pi/Desktop'
   # filename = paste(/home/pi/desktop)
#df.csv





```
